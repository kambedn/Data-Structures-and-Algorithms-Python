<h2>Implementing Stacks ADT using arrays</h2>

```python
class StackArray:
    __slots__ = '_data'
    def __init__(self):
        """Class constructor. O(1)"""
        self._data = []

    def __len__(self):
        """Allows to use len function on the class. It returns the number of elements in stack. O(1)"""
        return len(self._data)

    def isEmpty(self):
        """Returns True if the stack has no elements, False otherwise. O(1)"""
        return not self._data

    def push(self, e):
        """Adds an element e to the stack. It will become the top of the stack. O(1)"""
        self._data.append(e)

    def pop(self):
        """Removes and returns the element at the of the stack. O(1)"""
        if self.isEmpty():
            print('The stack is empty!')
        else:
            return self._data.pop()

    def top(self):
        """Returns the top element. O(1)"""
        if self.isEmpty():
            print('The stack is empty!')
        else:
            return self._data[-1]
            
```

```python
St = StackArray()
St.push(10)
St.push(20)
print(St.isEmpty())
print(len(St))
print(St.top())
print('Removed element: ', St.pop())
print(St.top())
St.push(30)
St.push(20)
print(St._data)
```

<h2>Using stack to match parantheses pattern</h2>
<h4>The goal of this task is to write a function which uses a StackArray object to determine 
    <br>whether a given string matches a pattern: {[([])]}<br></h4>

```python
def match_pattern(s):
    stack = StackArray()
    for char in s:
        if char in ['[', '(', '{']:
            stack.push(char)
        elif char in [']', ')', '}']:
            last = stack.top()
            if ( char == ']' and last == '[' ) or ( char == ')' and last == '(' ) or ( char == '}' and last == '{' ):
                stack.pop()
            else:
                break
        else:
            break
    else:
        if stack.isEmpty():
            return True
    return False
```

```python
print(match_pattern('{[([])]}'))
print(match_pattern('{[([])]}['))
print(match_pattern('{[([])]}[)'))
print(match_pattern('{[([])]}a[)'))
```

<h2>Implementing Stacks ADT using linked lists</h2>

```python
class _Node:
    __slots__ = '_element', '_next'
    def __init__(self, e, next):
        self._element = e
        self._next = next


class StackLinkedList():
    __slots__ = '_top', '_size'
    def __init__(self):
        """Class constructor. O(1)"""
        self._top = None
        self._size = 0

    def __len__(self):
        """Allows to use len function on the class. It returns the number of elements in stack. O(1)"""
        return self._size

    def isEmpty(self):
        """Returns True if the stack has no elements, False otherwise. O(1)""" 
        return not self._size

    def push(self, e):
        """Adds element e to the stack. It will become the top of the stack. O(1)"""
        new = _Node(e, None)
        if not self.isEmpty():
            new._next = self._top
        self._top = new
        self._size += 1

    def pop(self):
        """Removes and returns the top element. O(1)"""
        if self.isEmpty():
            print('The stack is empty!')
            return
        e = self._top._element
        self._top = self._top._next
        self._size -= 1
        return e

    def top(self):
        """Returns the top element. O(1)"""
        if self.isEmpty():
            print('The stack is empty!')
        else:
            return self._top._element
            
```

```python
stackl = StackLinkedList()
print(stackl.top())
stackl.push(10)
stackl.push(20)
stackl.push(30)
print(stackl.top())
print(len(stackl))
print(stackl.isEmpty())
print(stackl.pop())
print(stackl.pop())
print(stackl.pop())
print(stackl.isEmpty())
```
