<h1>Recursion and analaysis of recursive functions</h1>

<h5>Example #1: Iteration vs recursion - displaying square of N numbers</h5>

```python
def sqr_iter(n):
    while n>0:
        print(n*n)
        n -= 1


sqr_iter(6)
```

```python
def sqr_rec(n):
    if n>0:
        print(n*n)
        sqr_rec(n-1)


sqr_rec(6)
```

<h3><b>Types of recursion</b></h3>

<h5>#1 Tail recursion</h5>
When a function calls itself and that call statement is the last one in the function. <br>
There should be NO statements AFTER the recursive call.

```python
def calculate(n):
    if n>0:
        print(n*n)
        calculate(n-1)


calculate(4)
```

<h5>#2 Head recursion</h5>
When a function calls itself and that call statement is the first in the function.<br>
There should be NO statements (except for checking for the base case) BEFORE the recursive call.

```python
def calculate(n):
    if n>0:
        calculate(n-1)
        print(n*n)


calculate(4)
```

<b>THIS IS NOT A HEAD RECURSION</b><br>
Some computation is done after the recursive call.

```python
def calculate(n):
    if n>0:
        print(n*n)
        return calculate(n-1)+2
    else:
        return 0


print(calculate(4))
```

<h5>#3 Tree recursion</h5>
When a function calls itself more than once.

```python
def calculate(n):
    if n>0:
        calculate(n-1)
        print(n*n)
        calculate(n-1)


calculate(4)
```

<h5>#4 Indirect recursion</h5>

If we have more than one function and each one of them, depending upon the base conidiotn, <br>calls the other function and the last function also depending upon the base condition calls the first function.<br>
The functions create <b>a loop or a circular pattern</b>.<br>
In indirect recusion any function <b>does not call itself</b>.<br>
There should be a base case, so ultimately the recursive calls are terminated.

<h4>Sum of N natural numbers</h4>

Recursive approach - linear time complexity O(n)<br>
sum(n) = sum(n-1) + n <br>
sum(n) = sum(n-2) + n-1 + n etc.

```python
def sum(n):
    if n == 1: # base case
        return 1
    return sum(n-1) + n


print(sum(20))
```

Mathematical approach using formula - constant time complexit O(1)<br>
sum(n) = n*(n+1)/2

```python
def sum_f(n):
    return n*(n+1)/2


print(sum(20))
```

Iterative approach - linear time complexity O(n) <br>
Adding each number in a loop

```python
def sum_i(n):
    sum = 0
    for i in range(n+1):
        sum += i
    return sum


print(sum_i(20))
```

<h4>Factorial of a number</h4>

Recursive approach - linear time complexity O(n) <br>
fact(n) = fact(n-1) * n <br>
fact(n) = fact(n-2) * (n-1) * n etc.

```python
def factorial_r(n):
    if not n: # base case
        return 1
    return factorial_r(n-1)*n


print(factorial_r(6))
```

Iterative approach - linear time complexity O(n) <br>
Multiplying numbers in a loop

```python
def factorial_i(n):
    fact = 1
    for i in range(1, n+1):
        fact *= i
    return fact


print(factorial_i(6))
```
