<h1><b>Searching algorithms</b></h1>

<h5>A. Linear/Sequential search </h5>
The linear seach sequentially checks each element of the list until a match is found<br>
or the whole list has been searched. <br>
<b>Linear</b> time complexity.

```python
def linear_search(A, key):
    for i in range(len(A)):
        if A[i] == key:
            return i # returning the index
    return -1  # if the key is not found


A = [21, 45, 67, 42, 100]
print(linear_search(A, 22))
print(linear_search(A,42))
```

<h5>B. Binary search </h5>
<b>Logarithmic</b> time complexity. O(log n)

```python
def binary_search_iter(A, key):
    L = 0
    R = len(A) - 1
    while L <= R:
        m = (L+R)//2
        if A[m] == key:
            return m
        elif A[m] < key:
            L = m + 1
        else:
            R = m - 1
            
    return -1


A = [21, 45, 67, 42, 100]
A.sort()
print(A)
print(binary_search_iter(A, 22))
print(binary_search_iter(A,42))
```

<h5>Binary search using recursion</h5>
Also <b>logarithmic</b> time complexity. O(log n)

```python
def binary_search_rec(A,key,L=0,R=len(A)-1):
    if L > R: 
        return -1
    else:
        m = (L+R)//2
        if A[m] == key:
            return m
        elif A[m] < key:
            return binary_search_rec(A,key,L=m+1,R=R)
        else:
            return binary_search_rec(A,key,L=L, R=m-1)


A = [21, 45, 67, 42, 100]
A.sort()
print(A)
print(binary_search_rec(A, 22))
print(binary_search_rec(A,42))
```
